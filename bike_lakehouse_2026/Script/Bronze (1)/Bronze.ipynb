#Reading from csv file & Writing to Bronze layer

```python
df = (
        spark.read.option("header", "true")
        .option("inferschema", "true")
        .csv("/Volumes/workspace/bronze/source_system/source_crm/cust_info.csv")
    )
df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("bronze.cust_info")
df.limit(10).display()
```
